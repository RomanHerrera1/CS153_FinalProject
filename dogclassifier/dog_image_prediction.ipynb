```python
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.models import load_model
import os 
import zipfile 
import tensorflow as tf 
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
from tensorflow.keras import layers 
from tensorflow.keras import Model 
import matplotlib.pyplot as plt
 
```

```python
# load and prepare the image
def load_image(filename):
    # load the image
    img_to_show = load_img(filename)
    plt.imshow(img_to_show)
    
    img = load_img(filename, target_size=(150, 150))
    # convert to array
    img = img_to_array(img)
    # reshape into a single sample with 3 channels
    img = img.reshape(1, 150, 150, 3)
    # center pixel data
    img = img.astype('float32')
#     img = img - [123.68, 116.779, 103.939]
    img = img - [82.82, 78.2, 103.939]
    return img
 
# load an image and predict the class
def run_example():
    # load the image
    img = load_image('zorua1.jpg')
    # load model
    model = load_model('model.h5')
    # predict the class
    result = model.predict(img)
    
#     print(result[0])
    if result[0][0] == 0.0:
        print("The model predicts this image to be an artificial dog")
    elif result[0][0] == 1.0:
        print("The model predicts this image to be a real dog")
    elif result[0][0] < 0.5 and result[0][0] > 0:
        print("The model leans towards this image being an artificial dog, with a prediction score of ", result[0][0])
    elif result[0][0] >= 0.5 and result[0][0] < 1:
        print("The model leans towards this image being a real dog, with a prediction score of ", result[0][0])
    else:
        print("The model is unsure of if the image is a real dog or artificial dog")
 
# entry point, run the example
run_example()
```
